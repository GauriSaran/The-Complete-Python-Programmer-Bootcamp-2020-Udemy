# Simulating a random walk:
### IF we get a Head, we go step up direction and then travel horizontally in right direction.
### IF we get Tails, we go step down direction and then travel horizontally in right direction.

```python
from random import choice
step_option = [1,-1]
step_choice = choice(step_option)
step_choice
```

```python
walk = []
walk.append(step_choice)
for i in range(1, 1000):
    next_step = walk[i-1] + choice(step_option)
    walk.append(next_step)
walk
```

```python
plt.plot(walk)
```

# We can create a function for above code:

```python
def random_walk(step_num):
    step_option = [1,-1]
    step_choice = choice(step_option)
    walk = []
    walk.append(step_choice)
    for i in range(1, 1000):
        next_step = walk[i-1] + choice(step_option)
        walk.append(next_step)
    return walk
```

```python
rand_w = random_walk(1000)
print(len(rand_w))
```

```python
def plot_walk(walk):
    x = list(range(1,len(walk)+1))
    plt.plot(x,walk)
    plt.xlabel('Number of steps')
    plt.ylabel('Distance from origin')
    plt.title('Our random walk')
    plt.show()
```

```python
plot_walk(random_walk(1000))
```

# We can create many number of random walks and plot them on same plot
## num: is the number of random walks we want
## walk_steps: number of steps we want to walk per walk

```python
#plt.figure(figsize=(20,20))
def plot_walks(num, walk_steps):
    lab_list= list(range(1,num+1))
    for i in range(0,num):
        x = list(range(1,walk_steps+1))
        plt.plot(x, random_walk(walk_steps), label = 'Plot Number ' + str(lab_list[i]))
        plt.xlabel('Number of steps')
        plt.ylabel('Distance from origin')
        plt.legend(loc = 'lower left')
    
    plt.show()
```

```python
plot_walks(10,1000)
```

# Sierpinski Triangle

First Transformation 
\begin{equation}
x_{n+1} = 0.5x_n 
\end{equation}
\begin{equation}
y_{n+1} = 0.5y_n
\end{equation}
Second Transformation
\begin{equation}
x_{n+1} = 0.5x_n + 0.5
\end{equation}
\begin{equation}
y_{n+1} = 0.5y_n + 0.5
\end{equation}
Third Transformation
\begin{equation}
x_{n+1} = 0.5x_n + 1
\end{equation}
\begin{equation}
y_{n+1} = 0.5y_n
\end{equation}

```python
def trans_1(p):
    x = p[0]
    y = p[1]
    x1 = 0.5 * x
    y1 = 0.5 * y
    return x1,y1

def trans_2(p):
    x = p[0]
    y = p[1]
    x1 = 0.5 * x + 0.5
    y1 = 0.5 * y + 0.5
    return x1,y1

def trans_3(p):
    x = p[0]
    y = p[1]
    x1 = 0.5 * x + 1
    y1 = 0.5 * y
    return x1,y1


transformations = [trans_1,trans_2,trans_3]
a1 = [0]
b1 = [0]
a,b = 0,0
 
for i in range(10000000):
    trans = choice(transformations)
    a,b = trans((a,b))
    a1.append(a)
    b1.append(b)
```

```python
plt.rc('figure',figsize=(16,16))
```

```python
plt.plot(a1,b1,'o')
#plt.savefig('my_figure.png')
```
